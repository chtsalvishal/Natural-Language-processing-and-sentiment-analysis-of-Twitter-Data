This section is used to read all the English tweets store them in Entweet.

```python
def English_tweet(t):
    
    # Declare Temp objects.
    en_tweet=[]
    for j in t:  
        #removing 0 with space unicode values from tweets. eg: \u2069
        tweet_encoded = (str(j).encode('ascii', 'ignore')).decode("utf-8")
        
    # Classify the tweet and get only English Tweets.
        (l,g)=langid.classify(tweet_encoded)
        # Check if in English.
        if(l=='en'):
            tweet_encoded=tweet_encoded.lower()
            en_tweet.append(tweet_encoded)                     
    # Return List of English tweets.
    return(en_tweet)
```

## Reading the data from Excel sheets provided.

* In this section we read all the excel sheets and perform parsing so that all the extra spaces are removed from the file.


```python
# Reading the file from given in location.
excel_data = pd.ExcelFile('31131441.xlsx')

# Variable declarations.
data_sheet_Raw={}

# Loop to Read data from Excel Sheet.
for i in range(len(excel_data.sheet_names)):

    # Cleaning   
    df =excel_data.parse(i)
     
    df.dropna(axis=0,inplace=True,how='all')
    df.dropna(axis=1,inplace=True,how='all')
    if(df.columns[0]!= 'text'):
        df.reset_index(inplace=True)
        df.drop(axis=1,columns='index',inplace=True)
        df.columns= df.iloc[0]
        df.drop(index=0,inplace=True)
    entweet=English_tweet(df['text'])
    # Assignment to Dictionary.
    data_sheet_Raw[excel_data.sheet_names[i]]=entweet 
```

## Tokenization

```python
# For tokenization
from nltk.tokenize import RegexpTokenizer

# Unigram_Dict to store for all days.
unigram_tokens={}
# Regex tokenizer
tokenizer= RegexpTokenizer(r"[a-zA-Z]+(?:[-'][a-zA-Z]+)?")

# Using loop to traverse through Dictionary we previously created.
for k,v in data_sheet_Raw.items():
    token_list=[]
    # Using Each day key to find individual Tweets.
    for i in v:
            # Store these tokens.
            token_list.extend(tokenizer.tokenize(i))
        
    # Append to Day wise Unigram tokens.
    unigram_tokens[k]=token_list

# Display Tokens.    
unigram_tokens


```

## Stemming

```python
# Stemming.

# Initiation
stemmer = PorterStemmer()
final_tokens =[]
final_words={}

# Getting Stemmed Dict.
for k,v in Stop_Dict.items():
    # Using Stemmer package to stem the words.
    final_tokens=[stemmer.stem(w) for w in v ]
    # Store in final dictionary for further use.
    final_words[k]=final_tokens
```

## Remove less frequent words.

```python
# Function to remove 
def removeLessFreqWords(key,less_list):
    # check if w not in less frequency list.
    return (key, [w for w in final_words[key] if w not in less_list])
```

```python
# Remove less frequent words
from nltk.probability import * # freq dist
for k,v in final_words.items():    
    # Get the Frequency distribution of less frequent words.
    less_frequent=FreqDist(v)
    # Hapaxes provides the words that occur only once.
    lessFreqWords = set(less_frequent.hapaxes())  
    
    # TOkenized dictionary without Less Frequent Words.
    tokenized_Dict = dict(removeLessFreqWords(k,lessFreqWords) for k in final_words.keys())
```

## Thershold removal.
### Identify words that have :
    1.Thershold above 60 days or
    2.Thershold less 5 days.
    This function dynamically takes the value of Thershold and provides list.

```python
def thershold_calculator(thers,seg):
    #Check for Thersold and remove 60 days words.

    # Identify Words from corpus.
    Thers_word = list(chain.from_iterable([set(value) for value in tokenized_Dict.values()]))

    # Create FreqDistribution for words in documents.
    Ther_fd = FreqDist(Thers_word)

    # Get most common values.
    check_Ther=Ther_fd.most_common()
    Ther_set_list= []
    # Loop to check all the token's frequency
    # Store in a temporary list.
    for i in range(len(check_Ther)):
        # Word of the iterator
        word_i=check_Ther[i][0]
        # Count attached with the word tuple.
        count_i=check_Ther[i][1]
        # Length of word in iteration.
        length_i=len(word_i)

        # Matching conditions of document count > 60days
        # Length of words less than 3.
        if(count_i>thers and seg==0): # For 60 days or more
            Ther_set_list.append(word_i)
        elif(count_i<thers and seg==1): # For 5 days or less
            Ther_set_list.append(word_i)
            
    # Thershold updated list is converted to Set for performance.
    Ther_set_list=set(Ther_set_list)
    return Ther_set_list
```

## Master Token Dictionary.
### We will remove the Thershold words and get the final Master Dictionary.

```python
# Function to remove thershold words from main list.
def removeThershold(key,thershold_list):
    return (key,[w for w in tokenized_Dict[key] if w not in thershold_list])

# Master Dictionary tokens with perfectly removed conditions.
Master_Token_Dict = dict(removeThershold(k,thershold_calculator(60,0)) for k in tokenized_Dict.keys())
```
